9. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in
“London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall
reviews, privates from price, dorms from price, facilities and property description. 

```python
import selenium
import pandas as pd9
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
driver9=webdriver.Chrome()
```

```python
driver9.get("https://www.hostelworld.com/")
```

```python
search_city=driver9.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[1]/div/div[2]/input')
search_city.send_keys('London')
```

```python
select_city=driver9.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[2]/div/ul/li[2]/button')
select_city.click()
```

```python
search_button=driver9.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[5]/button[2]')
search_button.click()
```

```python
h_name=[]
hotel=driver9.find_elements(By.XPATH,'//div[@class="property-info-header"]/div[1]/span')
for s in hotel:
    h_name.append(s.text)
print(h_name)
len(h_name)
```

```python
h_distance=[]
distance=driver9.find_elements(By.XPATH,'//div[@class="property-distance"]/span[2]')
for t in distance:
    h_distance.append(t.text)
print(h_distance)
len(h_distance)
```

```python
h_ratings=[]
ratings=driver9.find_elements(By.XPATH,'//div[@class="property-info-header"]/div[2]/div[1]/div[1]/span[1]')
for u in ratings:
    h_ratings.append(u.text)
print(h_ratings)
len(ratings)
```

```python
h_reviews=[]
review=driver9.find_elements(By.XPATH,'//div[@class="review"]/span')
for v in review:
    h_reviews.append(v.text)
print(h_reviews)
len(h_reviews)
```

```python
h_Oreviews=[]
oreview=driver9.find_elements(By.XPATH,'//div[@class="property-info-header"]/div[2]/div[1]/div[1]/span[2]')
for w in oreview:
    h_Oreviews.append(w.text)
print(h_Oreviews)
len(h_Oreviews)
```

```python
h_Pprice=[]
private_price=driver9.find_elements(By.XPATH,'//div[@class="property-accommodation-prices"]/div[1]')
for a in private_price:
    h_Pprice.append(a.text)
print(h_Pprice)
len(h_Pprice)
```

```python
h_Dprice=[]
dorm_price=driver9.find_elements(By.XPATH,'//div[@class="property-accommodation-prices"]/div[2]')
for b in dorm_price:
    h_Dprice.append(b.text)
print(h_Dprice)
len(h_Dprice)
```

```python
h_url=[]
url=driver9.find_elements(By.XPATH,'//a[@class="property-card-container horizontal"]')
for x in url:
    h_url.append(x.get_attribute('href'))
len(h_url)
```

```python
h_facilities=[]
for z in h_url:
    driver9.get(z)
    time.sleep(5)
    try:
        facility=driver9.find_element(By.XPATH,'//div[@class="facilities-container"]/ul/li/ul/li')
        h_facilities.append(facility.text)
    except NoSuchElementException:
        h_facilities.append('Not Present')
len(h_facilities)
```

```python
h_descriptions=[]
for y in h_url:
    driver9.get(y)
    time.sleep(5)
    try:
        description=driver9.find_element(By.XPATH,'//div[@class="description-container"]/div[1]/div[2]')
        h_descriptions.append(description.text)
    except NoSuchElementException:
        h_descriptions.append('Not Present')
len(h_descriptions)
```

```python
#h_name=[] h_distance=[] h_ratings=[] h_reviews=[] h_Oreviews=[] h_Pprice=[] h_Dprice=[] h_facilities=[] h_descriptions=[]
df9=pd9.DataFrame({"Hotel Name":h_name,"Distance":h_distance,"Ratings":h_ratings,"Reviews":h_reviews,"Overviews":h_Oreviews,"Private Price":h_Pprice,"Dorm price":h_Dprice,"Facilities":h_facilities,"Description":h_descriptions})
df9
```
