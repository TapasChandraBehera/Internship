4. Scrape the details of trending repositories on Github.com.
Url = https://github.com/

A) Repository title
B) Repository description
C) Contributors count
D) Language used
Note: - From the home page you have to click on the trending option from Explore menu through code. 

```python
import selenium
from selenium import webdriver
import pandas as pd4
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
driver4=webdriver.Chrome()
```

```python
driver4.get("https://github.com/")
```

```python
trendingRepository_seemore=driver4.find_element(By.XPATH,'/html/body/div[1]/div[6]/div/div/div/div/div/div/main/div/div/div/feed-container/div[5]/turbo-frame/article/div/div/div[1]/header/a')
trendingRepository_seemore.click()
```

```python
r_title=[]
title=driver4.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]/a')
for i in title:
    r_title.append(i.text)
print(r_title)
len(r_title)
```

```python
r_links=[]
links=driver4.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]/a')
for x in links:
    r_links.append(x.get_attribute('href'))
print(r_links)
len(r_links)
```

```python
r_description=[]
for j in r_links:
    driver4.get(j)
    time.sleep(10)
    try:
        description=driver4.find_element(By.XPATH,'/html/body/div[1]/div[6]/div/main/turbo-frame/div/div/div/div[2]/div[2]/div/div[1]/div/div/p')
        r_description.append(description.text)
    except NoSuchElementException:
        r_description.append('Not Present')
print(r_description)        
len(r_description)
```

```python
r_contributors1=[]
for k1 in r_links:
    driver4.get(k1)
    time.sleep(10)
    try:
        contributors1=driver4.find_element(By.XPATH,'//div[@class="Layout-sidebar"]/div/div[5]/div/h2/a/span')
        r_contributors1.append(contributors1.text)
    except NoSuchElementException:
        r_contributors1.append('Not Present')
print(r_contributors1)        
len(r_contributors1)
```

```python
r_language=[]
for l in r_links:
    driver4.get(l)
    time.sleep(10)
    try:
        language=driver4.find_element(By.XPATH,'//div[@class="Layout-sidebar"]/div/div[6]/div/ul/li/a/span[1]')
        r_language.append(language.text)
    except NoSuchElementException:
        r_language.append('Not Present')
print(r_language)        
len(r_language)
```

```python
df4=pd4.DataFrame({"Repository_Title":r_title,"Description":r_description,"Contributors1":r_contributors1,"Language":r_language})
df4
```

```python
df4.to_csv('Trending_Repository',index=False)
```

to find the following details:

B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes

```python
import selenium
from selenium import webdriver
import pandas as pd7
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver7=webdriver.Chrome()
```

```python
driver7.get("https://www.imdb.com/search/title/?title_type=tv_series&sort=num_votes,desc")
```

```python
name=[]
s_name=driver7.find_elements(By.XPATH,'//div[@class="ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-be6f1408-9 srahg dli-title"]/a/h3')
for i in s_name:
    name.append(i.text)
print(name)
len(name)
```

```python
year_span=[] 
yearspan=driver7.find_elements(By.XPATH,'//div[@class="sc-be6f1408-7 iUtHEN dli-title-metadata"]/span[1]')
for j in yearspan:
    year_span.append(j.text)
print(year_span)
len(year_span)
```

```python
links=[]
link=driver7.find_elements(By.XPATH,'//div[@class="ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-be6f1408-9 srahg dli-title"]/a')
for z in link:
    links.append(z.get_attribute('href'))
print(links)
len(links)
```

```python
genre=[]
for k in links:
    driver7.get(k)
    time.sleep(10)
    try:
        genres=driver7.find_element(By.XPATH,'//section[@class="sc-69e49b85-4 ktjuZl"]/div[1]/div[2]/a/span')
        genre.append(genres.text)
    except NoSuchElementException:
        genre.append('Not Present')
print(genre)        
len(genre)
```

```python
runtime=[]
for l in links:
    driver7.get(l)
    time.sleep(10)
    try:
        runtimes=driver7.find_element(By.XPATH,'//div[@class="sc-69e49b85-0 jqlHBQ"]/ul/li[4]')
        runtime.append(runtimes.text)
    except NoSuchElementException:
        runtime.append('Not Present')
print(runtime)        
len(runtime)
```

```python
ratings=[]
for m in links:
    driver7.get(m)
    time.sleep(10)
    try:
        rating=driver7.find_element(By.XPATH,'//div[@class="sc-3a4309f8-0 bjXIAP sc-69e49b85-1 llNLpA"]/div/div[1]/a/span/div/div[2]/div[1]/span[1]')
        ratings.append(rating.text)
    except NoSuchElementException:
        ratings.append('Not Present')
print(ratings)        
len(ratings)
```

```python
votes1 = driver7.find_elements(By.XPATH, '//div[@class="sc-f24f1c5c-0 cPpOqU"]')
votes = []
for element in votes1:
    text = element.text.strip("Votes")
    votes.append(text)
print(votes)
len(votes)
```
