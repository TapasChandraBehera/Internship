1) Write a python program to display all the header tags from "https://en.wikipedia.org/wiki/Main_Page" and make data frame.

```python
from bs4 import BeautifulSoup
import requests
#Sending request to the webpage to get the source code
page1=requests.get("https://en.wikipedia.org/wiki/Main_Page")
#To get the page content
soup1=BeautifulSoup(page1.content)
#for scrapping or to display all the header tags from the website
headers=[]
for i in soup1.find_all('span',class_="mw-headline"):
    headers.append(i.text)
#To make a data frame we have to import pandas library
import pandas as pd
df1=pd.DataFrame({'Headers':headers})
df1  
```

2.Use this link "https://presidentofindia.nic.in/former-presidents" to extract all the former presidents names and make a dataframe.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
url = "https://presidentofindia.nic.in/former-presidents"
response = requests.get(url)
html_content = response.content
soup = BeautifulSoup(html_content, "html.parser")
president_names = []
for president_div in soup.find_all("div", class_="desc-sec"):
    name = president_div.find("h3").text
    president_names.append(name)
presidents_df = pd.DataFrame({"President Names": president_names})
print(presidents_df)
```

```python
#3
import requests
from bs4 import BeautifulSoup
import pandas as pd
url = requests.get('https://www.icc-cricket.com/rankings/team-rankings/mens/odi')  # Replace 'Your_URL' with the actual URL
soup=BeautifulSoup(url.content)
teams = []
rankings = []
matches = []
points = []
table_body = soup.find('div', class_='si-table-body')
for row in table_body.find_all('div', class_='si-table-row')[:10]:
    rank = row.find('div', class_='si-table-data si-pos').text.strip()
    team_name = row.find('div', class_='si-table-data si-team').find('span', class_='si-text').text.strip()
    match = row.find('div', class_='si-table-data si-matches').text.strip()
    point = row.find('div', class_='si-table-data si-pts').text.strip()
    # Append the extracted data to lists
    rankings.append(rank)
    teams.append(team_name)
    matches.append(match)
    points.append(point)
data = {'Ranking': rankings, 'Team': teams, 'Matches': matches, 'Points': points}
df = pd.DataFrame(data)
print(df)
```

5.Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame-
i) Headline
ii) Time
iii) News Link

```python
#5
from bs4 import BeautifulSoup
import requests
import pandas as pd
page5=requests.get("https://www.cnbc.com/world/?region=world")
soup5=BeautifulSoup(page5.content)
n_headline=[]
for i in soup5.find_all('a',class_="LatestNews-headline"):
    n_headline.append(i['title'])
n_time=[]
for i in soup5.find_all('span',class_="LatestNews-wrapper"):
    n_time.append(i.text)
n_link=[]
for i in soup5.find_all('a',class_="LatestNews-headline"):
    n_link.append(i['href'])
df5=pd.DataFrame({'Headline':n_headline,'Time':n_time,'News Link':n_link})
df5
```

6.Write a python program to scrape the details of most downloaded articles from AI in last 90 days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles Scrape below mentioned details and make data frame-
i) Paper Title
ii) Authors
iii) Published Date
iv) Paper URL

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
page6=requests.get("https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles")
soup6=BeautifulSoup(page6.content)
p_title=[]
for i in soup6.find_all('h2',class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
    p_title.append(i.text)
p_author=[]
for i in soup6.find_all('span',class_="sc-1w3fpd7-0 dnCnAO"):
    p_author.append(i.text)
p_pdate=[]
for i in soup6.find_all('span',class_="sc-1thf9ly-2 dvggWt"):
    p_pdate.append(i.text)
p_url=[]
for i in soup6.find_all('a',class_="sc-5smygv-0 fIXTHm"):
    p_url.append(i['href'])
df6=pd.DataFrame({'Paper_Title':p_title,'Author':p_author,'Published_Date':p_pdate,'Paper_URL':p_url})
df6
```

7.Write a python program to scrape the details of most downloaded articles from AI in last 90 days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles Scrape below mentioned details and make data frame-
i) Paper Title
ii) Authors
iii) Published Date
iv) Paper URL

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
page7=requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
soup7=BeautifulSoup(page7.content)
r_name=[]
for i in soup7.find_all('div',class_="restnt-info cursor"):
    r_name.append(i.text)
r_cuisine=[]
for i in soup7.find_all('span',class_="double-line-ellipsis"):
    r_cuisine.append(i.text)
r_location=[]
for i in soup7.find_all('div',class_="restnt-loc ellipsis"):
    r_location.append(i.text)
r_rating=[]
for i in soup7.find_all('div',class_="restnt-rating rating-4"):
    r_rating.append(i.text)
r_image=[]
for i in soup7.find_all('img',class_="no-img"):
    r_image.append(i['data-src'])
df7=pd.DataFrame({'Restaurants':r_name,'Cuisine':r_cuisine,'Location':r_location,'Rating':r_rating,'Image':r_image})
df7
```
